```python
import gradio as gr
import assemblyai as aai
import time
from PyPDF2 import PdfReader
from langchain.text_splitter import RecursiveCharacterTextSplitter

import os
from langchain_google_genai import GoogleGenerativeAIEmbeddings
import google.generativeai as genai  # google gemini pro
from langchain.vectorstores import FAISS
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.chains.question_answering import load_qa_chain  # RAG
from langchain.prompts import PromptTemplate
from dotenv import load_dotenv
import fitz  # PyMuPDF
import re
from gtts import gTTS
import requests
from bs4 import BeautifulSoup

load_dotenv()

# Set your AssemblyAI and Google API keys
aai.settings.api_key = "9ccc621903974aec9abc13255f93984a"
os.environ["GOOGLE_API_KEY"] = "AIzaSyBd4k2h_Lw1FZONTmCQZke82VsFVRsUBDE"
genai.configure(api_key=os.environ["GOOGLE_API_KEY"])

def transcribe_audio(file):
    transcriber = aai.Transcriber()
    transcript = transcriber.transcribe(file)

    # Polling for the transcription status
    while transcript.status not in [aai.TranscriptStatus.completed, aai.TranscriptStatus.error]:
        time.sleep(5)  # Wait for 5 seconds before polling again
        transcript = transcriber.get_transcript(transcript.id)

    if transcript.status == aai.TranscriptStatus.error:
        return transcript.error
    else:
        return transcript.text

def get_pdf_text(pdf_docs):
    text = ""
    for pdf in pdf_docs:
        pdf_reader = PdfReader(pdf)
        for page in pdf_reader.pages:
            text += page.extract_text()
    return text

def get_text_chunks(text):
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=10000,
        chunk_overlap=1000,
        length_function=len
    )
    chunks = text_splitter.split_text(text)
    return chunks

def get_url_text(url):
    response = requests.get(url)
    content = response.content
    soup = BeautifulSoup(content, 'html.parser')
    text = str(soup.get_text())
    text = re.sub(r'\n+', ' ', text)
    return text

def get_vector_store(text_chunks):
    embeddings = GoogleGenerativeAIEmbeddings(model="models/embedding-001")
    vector_store = FAISS.from_texts(text_chunks, embedding=embeddings)
    vector_store.save_local("faiss_index")

def get_conversational_chain():
    prompt_template = """
    Answer the question as detailed as possible from the provided context of about the college, make sure to provide all the college details. If the answer is not in
    the provided context, just say, "answer is not available in the context". Do not provide the wrong answer.

    Context:\n {context}?\n
    Question:\n {question}\n
    Answer:
    """
    model = ChatGoogleGenerativeAI(model="gemini-pro", temperature=0.3)
    prompt = PromptTemplate(template=prompt_template, input_variables=["context", "question"])
    chain = load_qa_chain(model, chain_type="stuff", prompt=prompt)
    return chain

def user_input(user_question):
    embeddings = GoogleGenerativeAIEmbeddings(model="models/embedding-001")
    new_db = FAISS.load_local("faiss_index", embeddings, allow_dangerous_deserialization=True)
    docs = new_db.similarity_search(user_question)
    chain = get_conversational_chain()
    response = chain({"input_documents": docs, "question": user_question}, return_only_outputs=True)
    return response["output_text"]

def process_input(input_area, pdf_docs, audio_file):
    text_chunks = []
    if len(input_area) > 300000:
        return "Text length exceeds 300000 words. Please enter a shorter text."

    if input_area:
        inputs = input_area.split()
        for input_text in inputs:
            if input_text.startswith('http'):
                url_text = get_url_text(input_text)
                text_chunks.extend(get_text_chunks(url_text))
            else:
                text_chunks.extend(get_text_chunks(input_text))
        get_vector_store(text_chunks)

    if pdf_docs:
        raw_text = get_pdf_text(pdf_docs)
        text_chunks.extend(get_text_chunks(raw_text))
        get_vector_store(text_chunks)

    if audio_file:
        audio_text = transcribe_audio(audio_file)
        if "error" in audio_text:
            return f"Error in audio transcription: {audio_text}"
        text_chunks.extend(get_text_chunks(audio_text))
        get_vector_store(text_chunks)

    return "Text, URLs, and PDF files processed successfully!"

def ask_question(user_question, user_audio):
    if user_audio:
        user_question = transcribe_audio(user_audio)
        if "error" in user_question:
            return f"Error in audio transcription: {user_question}", None

    text_audio = user_input(user_question)
    if text_audio:
        tts = gTTS(text=text_audio, lang='hi')
        tts.save("output.mp3")
        return text_audio, "output.mp3"
    else:
        return "No text to speak. Please enter a question.", None

input_area = gr.Textbox(lines=10, label="Enter text or URLs (max 300000 words)")
pdf_docs = gr.File(label="Upload PDF files", file_count="multiple", type="binary")
# audio_file = gr.Audio(label="Upload an audio file for processing", type="filepath")

inputs = [input_area, pdf_docs]
# inputs = [input_area, pdf_docs, audio_file]
outputs = gr.Textbox(label="Processed Status")

app = gr.Interface(fn=process_input, inputs=inputs, outputs=outputs, live=True)

user_question = gr.Textbox(label="Ask a Question")
user_audio = gr.Audio(label="Use Mic OR Upload an audio file with your question", type="filepath")
output_text = gr.Textbox(label="Reply")
output_audio = gr.Audio(label="Audio Reply")

app2 = gr.Interface(fn=ask_question, inputs=[user_question, user_audio], outputs=[output_text, output_audio])

demo = gr.TabbedInterface([app, app2], ["Process Input", "Ask Question"])

if __name__ == "__main__":
    demo.launch()
```
